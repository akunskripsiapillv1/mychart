```python
import os
import re

def preprocess_text_files(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Pola regex umum yang ingin diperbaiki
    patterns = [
        (r'\s*-\s*', '-'),  # rata -rata → rata-rata
        (r'\s+([.,:;!?])', r'\1'),  # spasi sebelum tanda baca → hapus
        (r'\(\s+', '('),  # buka kurung dengan spasi → (...
        (r'\s+\)', ')'),  # tutup kurung dengan spasi → ...)
    ]

    for filename in os.listdir(input_folder):
        if filename.endswith('.txt'):
            print(f"Preprocessing {filename} ...")
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            with open(input_path, 'r', encoding='utf-8') as f:
                text = f.read()

            # Terapkan semua pola perbaikan
            for pattern, replacement in patterns:
                text = re.sub(pattern, replacement, text)

            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(text)
```

```python
preprocess_text_files("indo-version/descriptions/", "indo-version-clean/descriptions/")
```

```python
preprocess_text_files("indo-version/titles/", "indo-version-clean/titles/")
```
